```python
def qa(df, name):
    print(f"\n--- {name} ---")
    print("shape:", df.shape)
    print("duplicates:", df.duplicated().sum())
    print("null % top 8:")
    print((df.isna().mean()*100).sort_values(ascending=False).head(8))

qa(orders, "orders")
qa(items, "items")
qa(payments, "payments")
```

```python
import pandas as pd

RAW = "../01_data_raw"

customers = pd.read_csv(f"{RAW}/olist_customers_dataset.csv")
products  = pd.read_csv(f"{RAW}/olist_products_dataset.csv")
sellers   = pd.read_csv(f"{RAW}/olist_sellers_dataset.csv")
reviews   = pd.read_csv(f"{RAW}/olist_order_reviews_dataset.csv")
cat       = pd.read_csv(f"{RAW}/product_category_name_translation.csv")

print("Loaded:", customers.shape, products.shape, sellers.shape, reviews.shape, cat.shape)
```

```python
customers.columns = customers.columns.str.lower().str.strip()
customers = customers.drop_duplicates()
customers.to_csv("../02_data_clean/customers_clean.csv", index=False)

# CATEGORY TRANSLATION
cat.columns = cat.columns.str.lower().str.strip()

# PRODUCTS (merge English category)
products.columns = products.columns.str.lower().str.strip()
products = products.drop_duplicates()
products = products.merge(cat, on="product_category_name", how="left")
products.to_csv("../02_data_clean/products_clean.csv", index=False)

# SELLERS
sellers.columns = sellers.columns.str.lower().str.strip()
sellers = sellers.drop_duplicates()
sellers.to_csv("../02_data_clean/sellers_clean.csv", index=False)

# REVIEWS (parse dates)
reviews.columns = reviews.columns.str.lower().str.strip()
for c in ["review_creation_date", "review_answer_timestamp"]:
    if c in reviews.columns:
        reviews[c] = pd.to_datetime(reviews[c], errors="coerce")
reviews = reviews.drop_duplicates()
reviews.to_csv("../02_data_clean/reviews_clean.csv", index=False)

print("Export done ✅")
```
